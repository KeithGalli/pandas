```python
#iloc or "integer location"

print('This is df.iloc[0:4]')
print('Print first 4 rows (0, 1, 2, 3), not inclusive of the "4":')
print(df.iloc[0:4])

print('--------------')
print('Read a Specific Location (R, C)')
print(df.iloc[2,1])

print('--------------')
print('Read Each Row')
for index, row in df.head(3).iterrows():
    print(index, row)
    #print(index, row['Name'])
    
# pandas.iterrows - Iterate over DataFrame rows as (index, Series) pairs.
```

```python
#ILoc by Pokemon Type

# # Only return Pokemon whose Type 1 == 'Fire'
df.loc[df['Type 1'] == "Fire"].tail()

df.loc[df['Type 1'] == "Dark"].tail()
```
